```python
import pandas as pd
import numpy as np
```

```python
dados_vendas = pd.read_json('/content/dados_vendas_clientes.json')
dados_vendas.head()
```

```python
dados_vendas = pd.json_normalize(dados_vendas['dados_vendas'])
dados_vendas.head()
```

```python
dados_locacao = pd.read_json('/content/dados_locacao_imoveis.json')
dados_locacao.head()
```

```python
dados_locacao = pd.json_normalize(dados_locacao['dados_locacao'])
dados_locacao.head()
#
```

# Transformações dos Dados I

- Remover os dados em listas dentro do DataFrame;
- Verificar os tipos de dados;
- Identificar colunas numéricas;
- Transformar a coluna numérica para o tipo numérico.

## dados_vendas

```python
dados_vendas
```

```python
colunas = list(dados_vendas.columns)
colunas
```

```python
dados_vendas = dados_vendas.explode(colunas[1:])
dados_vendas.head()
```

```python
dados_vendas.reset_index(inplace=True, drop=True)
dados_vendas.head()
```

```python
dados_vendas.info()
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','').strip())
```

```python
dados_vendas['Valor da compra'] = dados_vendas['Valor da compra'].astype(np.float64)
```

```python
dados_vendas.info()
```

## dados_locacao

```python
dados_locacao.head()
```

```python
colunas = list(dados_locacao.columns)
colunas
```

```python
dados_locacao = dados_locacao.explode(colunas[1:])
dados_locacao.head()
```

```python
dados_locacao.reset_index(inplace=True, drop=True)
dados_locacao.head()
```

```python
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].apply(lambda x: x.replace('$ ', '').replace(' reais', '').replace(',','.').strip())
```

```python
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].astype(np.float64)
```

```python
dados_locacao.info()
```

```python
dados_locacao.head()
```

# Transformações dos Dados II

## dados_vendas

```python
dados_vendas.head()
```

```python
dados_vendas['Cliente'] = dados_vendas['Cliente'].str.lower().replace('[^a-z ]', '', regex=True).str.strip()
```

```python
dados_vendas.head()
```

## dados_locacao

```python
dados_locacao.head()
```

```python
dados_locacao['apartamento'] = dados_locacao['apartamento'].apply(lambda x: x.replace('(blocoAP)','').strip())
```

```python
dados_locacao.head()
```

---------------------------------------------------------

```python
dados_vendas.head()
```

```python
dados_vendas['Data de venda'] = pd.to_datetime(dados_vendas['Data de venda'], format='%d/%m/%Y')
```

```python
dados_vendas.info()
```

```python
total_compras = dados_vendas.groupby(['Cliente'])['Valor da compra'].sum()
```

```python
total_compras
```

```python
dados_locacao.head()
```

```python
dados_locacao.info()
```

```python
dados_locacao['datas_combinadas_pagamento'] = pd.to_datetime(dados_locacao['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados_locacao['datas_de_pagamento'] = pd.to_datetime(dados_locacao['datas_de_pagamento'], format='%d/%m/%Y')
```

```python
dados_locacao.info()
```

```python
dados_locacao.head()
```

```python
dados_locacao['atraso'] = (dados_locacao['datas_de_pagamento'] - dados_locacao['datas_combinadas_pagamento']).dt.days
```

```python
dados_locacao.head()
```

```python
media_atraso = dados_locacao.groupby(['apartamento'])['atraso'].mean()
```
